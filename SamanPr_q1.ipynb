```
import numpy as np
def Q_1(M,N):
  """This function gives the value of P for one realization (experiment)"""
  #M = 2 ; N = 26
  lst_suits = [N]*M 
  lst_suits
  card1 = -1 ; card2 = -1 ; 
  score_lst = []
  while  sum(lst_suits) != 0:
    #print ('err')
    drawn_turn_idx1 = int(np.random.uniform(0,1,1) * (M))
    drawn_turn_idx2 = int(np.random.uniform(0,1,1) *(M))
    #print(drawn_turn_idx2)
    if lst_suits[drawn_turn_idx1] != 0:
      lst_suits[drawn_turn_idx1] -= 1 

    if lst_suits[drawn_turn_idx2] != 0:
      lst_suits[drawn_turn_idx2] -= 1
    
    if drawn_turn_idx2 == drawn_turn_idx1 and lst_suits[drawn_turn_idx2]>=0 and lst_suits[drawn_turn_idx1]>=0 : 
      score_lst.append(1)
    else : 
        score_lst.append(0)
  #print(lst_suits)
  return sum(score_lst)
def run_process(M,N,times=100):
  """this function simulates the experiments discussed in the question for number of times to make a realization 
  for Random variable P """
  P_lst = []
  for i in range(times):
    P_lst.append(Q_1(M,N))
  return (P_lst)
```
